```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re

nhl_df=pd.read_csv("data/nhl.csv")
cities=pd.read_html("data/wikipedia_data.htm")[1]
cities=cities.iloc[:-1,[0,3,5,6,7,8]]
```

```python
# Obtener la última palabra en los nombres de los equipos en el marco de datos NHL
nhl_df['team'] = nhl_df['team'].apply(lambda x: re.findall(r'\b\w+\b', x)[-1])

# Casos especiales con dos palabras
special_cases_nhl = {
    'Maple Leafs': 'Maple Leafs',
    'Red Wings': 'Red Wings',
    'Canadiens': 'Canadiens',
    'Senators': 'Senators',
    'Capitals': 'Capitals',
    'Penguins': 'Penguins',
    'Blue Jackets': 'Blue Jackets',
    'Devils': 'Devils',
    'Hurricanes': 'Hurricanes',
    'Islanders': 'Islanders',
    'Rangers': 'Rangers',
    'Flyers': 'Flyers',
    'Predators': 'Predators',
    'Jets': 'Jets',
    'Blues': 'Blues',
    'Stars': 'Stars',
    'Wild': 'Wild',
    'Avalanche': 'Avalanche',
    'Blackhawks': 'Blackhawks',
    'Golden Knights': 'Golden Knights',
    'Flames': 'Flames',
    'Oilers': 'Oilers',
    'Sharks': 'Sharks',
    'Ducks': 'Ducks',
    'Canucks': 'Canucks',
    'Kings': 'Kings',
    'Coyotes': 'Coyotes'
    # Agrega más casos según sea necesario
}

# Manejar casos especiales
for old_name, new_name in special_cases_nhl.items():
    nhl_df.loc[nhl_df['team'] == old_name, 'team'] = new_name

```
